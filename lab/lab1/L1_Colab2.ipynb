### Task 2.1

Plot a chart using that includes both temperature unit in the code box below:

```python
#Insert your coding below this line
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv('temperature_data.csv')

# Convert timestamp to datetime
df['time'] = pd.to_datetime(df['Timestamp'])

# Convert time to seconds from start
time_sec = (
    df['time'].dt.hour * 3600 +
    df['time'].dt.minute * 60 +
    df['time'].dt.second
)
t = time_sec - time_sec.iloc[0]

# Extract temperatures
temp_c = df['Temperature(C)']
temp_f = df['Temperature(F)']

# Create figure and axis
fig, ax1 = plt.subplots()

ax1.set_title('Temperature vs Time')
ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('Temperature (C)', color='tab:blue')

# Plot Celsius
ax1.plot(t, temp_c, '-o', color='tab:blue', label='Temperature (C)')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create second y-axis for Fahrenheit
ax2 = ax1.twinx()
ax2.set_ylabel('Temperature (F)', color='tab:red')

# Plot Fahrenheit
ax2.plot(t, temp_f, '-s', color='tab:red', label='Temperature (F)')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Grid for readability
ax1.grid(True)

plt.show()
```
