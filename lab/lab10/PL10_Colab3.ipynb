Sensor layer: The stethoscope captures structure-borne sound from the vacuum pump surface and feeds it into the USB microphone, which digitizes it at 48 kHz with 16-bit depth.
Server (MTConnect middleware 1): A headless server computer runs the MTConnect adapter, which samples the digital audio in 2048-point chunks and streams them as DisplacementTimeSeries data via the MTConnect agent over HTTP.
Raspberry Pi (MTConnect middleware 2): A collector script polls the MTConnect agent, parses the XML stream, and feeds the sound chunks into the trained CNN model. The CNN classifies each window into one of three states: OFF, ON-vacuum, or ON-airleaking. It also computes the sound level in dB.
Database: Both the operational state (Event) and sound level (Sample, dB) are inserted into MySQL with timestamps.
Visualization: Grafana queries MySQL and displays a time-series panel for sound level and a state timeline panel for operational state.

````python
### If you code a Python script for Task 3.4, please copy and paste your code here.
import requests
import numpy as np
from xml.etree import ElementTree as ET

## MTConnect info
agent = "mepotrb16.ecn.purdue.edu"
agent_port = "5001"
MTCONNECT_STR = "{urn:mtconnect.org:MTConnectStreams:1.4}"

N_SEQ = 24        # number of sequences for ~1 second (24 x 2048/48000 = 1.024s)
N_CHUNK = 2048    # samples per sequence

def get_current_sequence():
    """Get the latest sequence number from the agent."""
    url = f"http://{agent}:{agent_port}/current"
    response = requests.get(url)
    root = ET.fromstring(response.content)
    header = root.find(f".//{MTCONNECT_STR}Header")
    return int(header.attrib["lastSequence"])

def fetch_sound_data(from_seq, count):
    """Fetch 'count' sequences starting from from_seq."""
    url = f"http://{agent}:{agent_port}/sample?from={from_seq}&count={count}"
    response = requests.get(url)
    root = ET.fromstring(response.content)

    all_samples = []
    for ts in root.iter(f"{MTCONNECT_STR}DisplacementTimeSeries"):
        raw = ts.text.strip().split()
        samples = [int(x) for x in raw]
        all_samples.extend(samples)
    return np.array(all_samples, dtype=np.int16)

def get_one_second_data():
    """Collect ~1 second of sound data and return as numpy array."""
    last_seq = get_current_sequence()
    from_seq = last_seq - N_SEQ + 1  # go back N_SEQ sequences from latest
    sound_data = fetch_sound_data(from_seq, N_SEQ)
    print(f"Collected {len(sound_data)} samples ({len(sound_data)/48000:.3f} seconds)")
    return sound_data

def preprocess_for_cnn(sound_data):
    """Normalize and reshape for CNN input."""
    # Normalize to [-1, 1]
    sound_normalized = sound_data.astype(np.float32) / 32768.0
    # Reshape to (1, samples, 1) for CNN input
    cnn_input = sound_normalized.reshape(1, -1, 1)
    return cnn_input

## ============== MAIN ==============
if __name__ == "__main__":
    # Load your trained CNN model
    import tensorflow as tf
    model = tf.keras.models.load_model('your_cnn_model.h5', compile=False)

    label_map = {0: 'OFF', 1: 'ON-vacuum', 2: 'ON-airleaking'}

    # Fetch 1 second of data
    sound_data = get_one_second_data()

    # Preprocess
    cnn_input = preprocess_for_cnn(sound_data)

    # Predict
    prediction = model.predict(cnn_input)
    predicted_class = np.argmax(prediction, axis=1)[0]
    confidence = np.max(prediction)

    print(f"Predicted state: {label_map[predicted_class]} (confidence: {confidence:.4f})")
````
