Model improvements:
The CNN could be improved by training with more data, especially capturing more diverse examples of each state under different conditions (varying pump load, temperature, microphone placement). Using a deeper architecture with more convolutional layers or trying established architectures like ResNet or VGG adapted for audio could improve feature extraction. Adding dropout and batch normalization helps prevent overfitting to training conditions that may not generalize well in real time.
Post-processing the predictions:
The simplest fix without retraining is to apply a majority voting or smoothing algorithm on consecutive predictions. For example, instead of trusting a single inference, collect the last 5 predictions and only change the displayed state if the majority agree. This eliminates one-off false readings without changing the model at all.
Training data quality:
The Mel-spectrogram features may be capturing noise or transient sounds that don't represent the true steady-state of the pump. Recording longer, cleaner samples and using data augmentation (adding background noise, pitch shifting) during training makes the model more robust to real-world variation.
Threshold tuning:
Instead of always taking argmax, apply a confidence threshold — only accept a prediction if its probability exceeds, say, 0.8. If no class exceeds the threshold, keep the previous state. This prevents low-confidence switches between similar-sounding classes.
Feature engineering:
Air leaking and Vacuuming may produce similar low-frequency components that are hard to distinguish in Mel-spectrograms alone. Adding time-domain features like RMS energy or zero-crossing rate alongside the spectrogram could give the model more discriminating information.
